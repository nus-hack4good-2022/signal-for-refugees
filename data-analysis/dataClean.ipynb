```python
import pandas as pd
import numpy as np
refugeeData = pd.read_csv('population.csv')

# Remove the useless columns
refugeeData = refugeeData.drop(refugeeData.columns[[1,2,3,4]],axis=1)

# To get rid of all the years less than 1999
for i in range(0,len(refugeeData)):
    if refugeeData.loc[i,"Year"] < 1999:
        refugeeData = refugeeData.drop([i])
    else:
        pass

# Replace all NaN values with 0
refugeeData = refugeeData.replace(np.nan,0)

# Reset the row index and print dataframe
refugeeData = refugeeData.reset_index()
refugeeData = refugeeData.drop(['index'],axis=1)
refugeeData
```
